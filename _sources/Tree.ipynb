## 2. Entropi Dataset Awal (Parent Node)

### Distribusi Kelas dalam Dataset Awal

* Total data: 15
* Kelas Y: 6
* Kelas Z: 9

### Probabilitas Tiap Kelas

* $p_A = \frac{6}{15} = 0.4$
* $p_B = \frac{9}{15} = 0.6$



### Entropi Parent Node

$$
Entropy(S) = - (p_A \cdot \log_2 p_A + p_B \cdot \log_2 p_B)
$$

$$
= - (0.4 \cdot \log_2 0.4 + 0.6 \cdot \log_2 0.6)
$$

Gunakan:

* $\log_2 0.4 \approx -1.322$
* $\log_2 0.6 \approx -0.737$

$$
Entropy(S) = - (0.4 \cdot -1.322 + 0.6 \cdot -0.737)
$$

$$
= - (-0.5288 - 0.4422)
$$

$$
= 0.971
$$

### Hitung Entropi Left Node

$$
p_Y = \frac{1}{9} \approx 0.111, \quad p_Z = \frac{8}{9} \approx 0.889
$$

$$
Entropy = - (p_Y \cdot \log_2 p_Y + p_Z \cdot \log_2 p_Z)
$$

$$
= - (0.111 \cdot \log_2 0.111 + 0.889 \cdot \log_2 0.889)
$$

Gunakan:

* $\log_2 0.111 \approx -3.17$

$$
Entropy = - (0.111 \cdot -3.17 + 0.889 \cdot -0.17)
$$

$$
= - (-0.352 + -0.151) = 0.352 + 0.151 = {0.503}
$$



### Entropi untuk Right Node (Panjang > 235)

 Ambil Data dengan Panjang > 235:

| Panjang (cm) | Kelas |

| 250          | Y     |
| 237          | Z     |
| 247          | Y     |

Total data = 7

Kelas Y = 5

Kelas Z = 2



###  Hitung Entropi Right Node:

$$
p_Y = \frac{5}{7} \approx 0.714,\quad p_Z = \frac{2}{7} \approx 0.286
$$

$$
Entropy = - (p_Y \cdot \log_2 p_Y + p_Z \cdot \log_2 p_Z)
$$

Gunakan:

* $\log_2 0.714 \approx -0.485$
* $\log_2 0.286 \approx -1.807$

$$
Entropy = - (0.714 \cdot -0.485 + 0.286 \cdot -1.807)
$$

$$
= - (-0.346 + -0.517) = 0.346 + 0.517 = {0.863}
$$



### Menghitung Information Gain

###  Diketahui:

* Entropy(Parent) = 0.971
* Entropy(Left) = 0.503 (9 data)
* Entropy(Right) = 0.863 (6 data)


###  Rumus Information Gain:

$$
Gain = Entropy(S) - \left( \frac{|S_{left}|}{|S|} \cdot Entropy(S_{left}) + \frac{|S_{right}|}{|S|} \cdot Entropy(S_{right}) \right)
$$

$$
Gain = 0.971 - \left( \frac{9}{15} \cdot 0.503 + \frac{6}{15} \cdot 0.863 \right)
$$



### Hitung:

* $\frac{9}{15} \cdot 0.503 = 0.302$
* $\frac{6}{15} \cdot 0.863 = 0.345$

$$
Gain = 0.971 - (0.302 + 0.345) = 0.971 - 0.647 = {0.324}
$$

###  Entropi Left Node (Berat ≤ 215)

$$
p_Y = \frac{1}{7} \approx 0.143,\quad p_Z = \frac{6}{7} \approx 0.857
$$

$$
Entropy = - (p_Y \cdot \log_2 p_Y + p_Z \cdot \log_2 p_Z)
$$

Gunakan:

* $\log_2 0.143 \approx -2.81$
* $\log_2 0.857 \approx -0.22$

$$
Entropy = - (0.143 \cdot -2.81 + 0.857 \cdot -0.22)
= 0.402 + 0.188 = {0.590}
$$





### Entropi untuk Right Node (Berat > 215)


| No | Berat | Kelas |

###  Hitung Entropi:

$$
p_Y = \frac{5}{8} = 0.625,\quad p_Z = \frac{3}{8} = 0.375
$$

$$
Entropy = - (p_Y \cdot \log_2 p_Y + p_Z \cdot \log_2 p_Z)
$$

Gunakan:

* $\log_2 0.625 \approx -0.678$
* $\log_2 0.375 \approx -1.415$

$$
Entropy = - (0.625 \cdot -0.678 + 0.375 \cdot -1.415)
= 0.424 + 0.531 = {0.955}
$$



### Information Gain dari Split Berat ≤ 215

 Diketahui:

* Entropy Parent (sebelumnya):
  Kelas Y = 6, Kelas Z = 9

  $$
  p_Y = \frac{6}{15},\quad p_Z = \frac{9}{15}
  $$

  $$
  Entropy(S) = - (0.4 \cdot \log_2 0.4 + 0.6 \cdot \log_2 0.6) = {0.971}
  $$

* Entropy Left = 0.590 (7 data)

* Entropy Right = 0.955 (8 data)



### Hitung Information Gain:

$$
Gain = Entropy(S) - \left( \frac{7}{15} \cdot 0.590 + \frac{8}{15} \cdot 0.955 \right)
$$

* $\frac{7}{15} \cdot 0.590 = 0.275$
* $\frac{8}{15} \cdot 0.955 = 0.509$

$$
Gain = 0.971 - (0.275 + 0.509) = 0.971 - 0.784 = {0.187}
$$




### Hasil Perhitungan :

| Split Atribut     | Entropy Left | Entropy Right | Information Gain |
| ----------------- | ------------ | ------------- | ---------------- |
| **Panjang ≤ 235** | 0.503        | 0.863         | **0.324**        |
| **Berat ≤ 215**   | 0.590        | 0.955         | **0.187**        |




### Interpretasi

- Information Gain  menunjukkan seberapa besar "keuntungan informasi" yang diperoleh dari suatu pemisahan (split).
* Semakin tinggi  nilainya, semakin baik split tersebut dalam memisahkan kelas.



### Kesimpulan Akhir

- Split terbaik adalah pada atribut Panjang ≤ 235 dengan Information Gain = 0.324
- Split pada Berat ≤ 215  hanya menghasilkan Information Gain = 0.187

Jadi fitur Panjang lebih informatif  dan lebih layak dijadikan sebaga- node pertama (root)  dalam membangun decision tree.


## Visualisasi Decision Tree

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.preprocessing import LabelEncoder

# === Data terbaru ===
data = {
    "Panjang (cm)": [232, 249, 252, 240, 243, 226, 218, 212, 221, 219, 211, 250, 237, 213, 247],
    "Berat (kg)":   [219, 231, 214, 224, 218, 217, 211, 206, 203, 209, 201, 223, 227, 204, 220],
    "Kelas": ['Y', 'Y', 'Y', 'Z', 'Y', 'Z', 'Z', 'Z', 'Z', 'Z', 'Z', 'Y', 'Z', 'Z', 'Y']
}

df = pd.DataFrame(data)

# === Encode label Y/Z jadi 0/1 ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # Y = 1, Z = 0 (urutan alfabet)

# === Fitur dan Target ===
X = df[["Panjang (cm)", "Berat (kg)"]]
y = df["Label"]

# === Train Decision Tree ===
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3, random_state=42)
clf.fit(X, y)

# === Tampilkan aturan pohon (rules) ===
print("\n=== Aturan Pohon Keputusan ===")
tree_rules = export_text(clf, feature_names=list(X.columns))
print(tree_rules)

# === Visualisasi Pohon Keputusan ===
plt.figure(figsize=(12, 6))
plot_tree(clf,
          feature_names=["Panjang (cm)", "Berat (kg)"],
          class_names=le.classes_,  # ['Z', 'Y']
          filled=True,
          rounded=True,
          fontsize=10)
plt.title("Visualisasi Pohon Keputusan (Y vs Z)")
plt.show()
```

## Visualisasi Splitting Setelah ada 2 Data Baru

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.preprocessing import LabelEncoder

# === Data Y dan Z ===
data = {
    "Panjang (cm)": [232, 249, 252, 240, 243, 226, 218, 212, 221, 219, 211, 250, 237, 213, 247],
    "Berat (kg)":   [219, 231, 214, 224, 218, 217, 211, 206, 203, 209, 201, 223, 227, 204, 220],
    "Kelas": ['Y', 'Y', 'Y', 'Z', 'Y', 'Z', 'Z', 'Z', 'Z', 'Z', 'Z', 'Y', 'Z', 'Z', 'Y']
}

df = pd.DataFrame(data)

# === Encode label Y/Z jadi 0/1 ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # Z = 0, Y = 1

# === Fitur dan Target ===
X = df[["Panjang (cm)", "Berat (kg)"]]
y = df["Label"]

# === Train Decision Tree ===
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3, random_state=42)
clf.fit(X, y)

# === Prediksi 2 Data Baru ===
new_data = pd.DataFrame({
    "Panjang (cm)": [242, 213],
    "Berat (kg)": [229, 207]
})
preds = clf.predict(new_data)
predicted_labels = le.inverse_transform(preds)

# === Visualisasi Decision Boundary ===
x_min, x_max = X["Panjang (cm)"].min() - 5, X["Panjang (cm)"].max() + 5
y_min, y_max = X["Berat (kg)"].min() - 5, X["Berat (kg)"].max() + 5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 1),
                     np.arange(y_min, y_max, 1))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
plt.scatter(X["Panjang (cm)"], X["Berat (kg)"], c=y, cmap=plt.cm.coolwarm, edgecolor="k", s=60, label="Data Latih")
plt.scatter(new_data["Panjang (cm)"], new_data["Berat (kg)"], c="yellow", edgecolor="black", s=100, label="Data Baru")
for i, label in enumerate(predicted_labels):
    plt.text(new_data.iloc[i, 0]+0.5, new_data.iloc[i, 1], label, fontsize=10, weight='bold')

plt.xlabel("Panjang (cm)")
plt.ylabel("Berat (kg)")
plt.title("Decision Boundary dengan Data Baru")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print("\n=== Prediksi untuk Data Baru ===")
print(new_data.assign(Kelas_Prediksi=predicted_labels))
```
